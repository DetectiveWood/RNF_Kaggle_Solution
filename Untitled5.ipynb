```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

test = pd.read_csv("test.csv")
train = pd.read_csv("train.csv")
```

```python
train
```

```python
category = test.loc[:, train.dtypes == object].columns
category
```

```python
test = test.drop(columns=test.columns[6])
train = train.drop(columns=["Alley"])
test = test.drop(columns=['PoolQC'])
train = train.drop(columns=["PoolQC"])
test = test.drop(columns=['Fence'])
train = train.drop(columns=["Fence"])
test = test.drop(columns=['MiscFeature'])
train = train.drop(columns=["MiscFeature"])
test = test.drop(columns=['Utilities'])
train = train.drop(columns=["Utilities"])
test = test.drop(columns=['LandSlope'])
train = train.drop(columns=["LandSlope"])
test = test.drop(columns=['Street'])
train = train.drop(columns=["Street"])
test = test.drop(columns=['Condition2'])
train = train.drop(columns=["Condition2"])
```

```python
from sklearn import preprocessing
# Encode Categorical Columns
# создание и обучение
arr = ['MSZoning', 'LotShape', 'LandContour',
       'LotConfig','Neighborhood', 'Condition1',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual',
       'GarageCond', 'PavedDrive',
       'SaleType', 'SaleCondition']

le = preprocessing.LabelEncoder()
train['SaleType'] = le.fit_transform(train['SaleType'])

for i in arr:
    train[i] = le.fit_transform(train[i])
    test[i] = le.fit_transform(test[i])

train.head()
```

```python
train = train.fillna(train.mean())
test = test.fillna(test.mean())
```

```python
X = train.drop(columns=['SalePrice'])
y = train['SalePrice']
```

```python
from sklearn.model_selection import train_test_split

train = (train-train.min ())/ (train.max () - train.min ())
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(max_depth=7, min_samples_leaf=2, n_estimators=50, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
y_pred_test = model.predict(test)

y_pred_test = pd.DataFrame(data = y_pred_test, columns=['SalePrice'])

y_pred_test = y_pred_test.reset_index()
```

```python
from sklearn import metrics
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
for i in range(460):
    y_pred_test['index'] = y_pred_test['index'].replace(i, i+1001)
```

```python
y_pred_test.to_csv('solution.csv', index = False)
```
